```python
#%% Inicialización
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import patches
from matplotlib.figure import Figure
from matplotlib import rcParams
import scipy.signal as sig

#%% Generación de señales de prueba

def zplane(b,a,filename=None):
    """
    Plot the complex z-plane given a transfer function.
    """

    # get a figure/plot
    ax = plt.subplot(111)

    # create the unit circle
    uc = patches.Circle((0,0), radius=1, fill=False,
                        color='black', ls='dashed')
    ax.add_patch(uc)

    # The coefficients are less than 1, normalize the coeficients
    if np.max(b) > 1:
        kn = np.max(b)
        b = b/float(kn)
    else:
        kn = 1

    if np.max(a) > 1:
        kd = np.max(a)
        a = a/float(kd)
    else:
        kd = 1
        
    # Get the poles and zeros
    p = np.roots(a)
    z = np.roots(b)
    k = kn/float(kd)
    
    # Plot the zeros and set marker properties    
    t1 = plt.plot(z.real, z.imag, 'go', ms=10)
    plt.setp( t1, markersize=10.0, markeredgewidth=1.0,
              markeredgecolor='k', markerfacecolor='g')

    # Plot the poles and set marker properties
    t2 = plt.plot(p.real, p.imag, 'rx', ms=10)
    plt.setp( t2, markersize=12.0, markeredgewidth=3.0,
              markeredgecolor='r', markerfacecolor='r')

    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    # set the ticks
    r = 1.5; plt.axis('scaled'); plt.axis([-r, r, -r, r])
    ticks = [-1, -.5, .5, 1]; plt.xticks(ticks); plt.yticks(ticks)

    if filename is None:
        plt.show()
    else:
        plt.savefig(filename)
    

    return z, p, k

def plot_module_and_phase(ww, hh):

    # Creo una figura y un subgráfico
    fig, ax1 = plt.subplots()

    # Escribo título del gráfico
    ax1.set_title('Respuesta en frecuencia filtro FIR')

    # Grafico respuesta de módulo
    ax1.plot(ww, 20 * np.log10(abs(hh)), color='b')

    # Escribo título eje y
    ax1.set_ylabel('Amplitud [dB]', color='b')

    # Escribo título eje x
    ax1.set_xlabel('Frecuencia normalizada')

    # Genero segundo subgráfico
    ax2 = ax1.twinx()

    # Grafico respuesta de fase
    ax2.plot(ww, np.unwrap(np.angle(hh)), color='g')

    # Agrego título eje y
    ax2.set_ylabel('Angulo [rad]', color='g')

    # Dibujo la grilla
    ax2.grid()

    # Muestro la figura
    plt.show()
```

```python
##########################################
# Acá podés generar los gráficos pedidos #
##########################################

# Función transferencia: N = 3
k = (1/3)
b = np.array([1, 1, 1])
a = np.array([1, 0, 0])

# Calculo la respuesta en frecuencia
[ww, hh] = sig.freqz(k*b, a)

# Normalizo el vector de frecuencias
ww = ww / np.pi

# Gráfico de respuesta de módulo y fase #
plot_module_and_phase(ww, hh)

# Grafico diagrama de polos y ceros #
zplane(k*b, a)
```

<div class="alert alert-block alert-warning">
<b>Respuesta:</b> Estudiando las señales ECG, se busca que los filtros utilizados reduzcan la interferencia de red en la señal promediada y que se eliminen los transitorios antes y despues del complejo QRS, pues un transitorio podría confundirse con los potenciales ventriculares tardíos.
    Para filtrar este tipo de ruido se utilizan filtros paso-alto con frecuencia de corte entre 0.5 y 1 Hz para eliminar actividades de muy baja frecuencia tales como las señales electrogalvánicas de los electrodos o artefactos de movimiento. También se debe aplicar un filtro paso-bajo con frecuencia de corte entre 35 y 70 Hz para evitar artefactos de elevada frecuencia como la actividad muscular.
       Tomando como piso de ruido 100mV, se decide generar un filtro pasa-bajo que atenúe por debajo del piso de ruido las señales de frecuencia superiores a 45 Hz (teniendo presente la ocasional deriva de la frecuencia de red); y observando el espectro del complejo QRS, se decide eliminar todas las componentes inferiores a 8 Hz (primera frecuencia de interés).
    Los filtros FIR se diseñan utilizando la técnica equiripple, con devío máximo admisible en banda de paso de 100mV.
</div>

```python
#%% Inicialización
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.signal as sig
import scipy.fftpack as sc
import scipy.io

#%% Generación de señales de prueba

# Cargo el registro electrocardiográfico (ECG)
mat = scipy.io.loadmat('ECG_TP4.mat')

ecg_lead = mat['ecg_lead']                                          # Registro de ECG muestreado a 𝑓𝑠=1  KHz durante una prueba de esfuerzo
qrs_pattern1 = mat['qrs_pattern1']                                  # Complejo de ondas QRS normal
heartbeat_pattern1 = mat['heartbeat_pattern1']                      # Latido normal
heartbeat_pattern2 = mat['heartbeat_pattern2']                      # Latido de origen ventricular
qrs_detections = mat['qrs_detections']                              # Vector con las localizaciones (en # de muestras) donde ocurren los latidos

# Frecuencia/periodo de muestreo
fs = 1000                                                           # Frecuencia de muestreo (en Hz)
ts = 1/fs                                                           # Periodo de muestreo 
```

```python
# Señal ECG
ecg_N = 10000                                                       # Cantidad de muestras ecg a graficar

ecg_tt = np.linspace(0, (ecg_N-1)*ts, ecg_N).flatten()              # Vector de sampleo temporal
ecg_spectrum = (2/ecg_N)*np.abs(sc.fft(ecg_lead[:ecg_N].flatten())) # Espectro de la señal ECG
ecg_half_spectrum = ecg_spectrum[:ecg_N//2]                         # Parte util del espectro de la señal ECG
ecg_ff = np.linspace(0, fs/2, ecg_N/2).flatten()                    # Vector de sampleo en frecuecia

#%% Generación de señales de prueba

# Registro ECG en tiempo
plt.figure(1)
plt.title('Registro electrocardiográfico ECG')
plt.xlabel('Tiempo [seg]')
plt.ylabel('Amplitud [mV]')
plt.grid(which='both', axis='both')
plt.plot(ecg_tt, ecg_lead[:ecg_N], color='b')
plt.show()

# Registro ECG en frecuencia
plt.figure(2)
plt.title('Espectro registro electrocardiográfico ECG')
plt.xlabel('Frecuecnia [Hz]')
plt.ylabel('Amplitud [mV]')
plt.grid(which='both', axis='both')
plt.stem(ecg_ff, ecg_half_spectrum)
plt.show()
```

```python
# Señal QRS
qrs_N = len(qrs_pattern1)

qrs_tt = np.linspace(0, (qrs_N-1)*ts, qrs_N).flatten()                  # Vector de sampleo temporal señal QRS
qrs_spectrum = (2/qrs_N)*np.abs(sc.fft(qrs_pattern1[:qrs_N].flatten()))
qrs_half_spectrum = qrs_spectrum[:qrs_N//2]
qrs_ff = np.linspace(0, fs/2, qrs_N/2)

# Registro QRS en tiempo
plt.figure(3)
plt.title('Complejo QRS')
plt.xlabel('Tiempo [seg]')
plt.ylabel('Amplitud [mV]')
plt.grid(which='both', axis='both')
plt.plot(qrs_tt, qrs_pattern1[:qrs_N], color='b')
plt.show()

# Registro QRS en frecuncia
plt.figure(4)
plt.title('Espectro complejo QRS')
plt.xlabel('Frecuecnia [Hz]')
plt.ylabel('Amplitud [mV]')
plt.stem(qrs_ff, qrs_half_spectrum)
plt.grid(which='both', axis='both')
plt.show()
```

```python
# Señal heartbeat_pattern1 y heartbeat_pattern2
hb_p1_N = len(heartbeat_pattern1)
hb_p2_N = len(heartbeat_pattern2)

hb_p1_tt = np.linspace(0, (hb_p1_N-1)*ts, hb_p1_N).flatten()
hb_p2_tt = np.linspace(0, (hb_p2_N-1)*ts, hb_p2_N).flatten()

hb_p1_spectrum = (2/hb_p1_N)*np.abs(sc.fft(heartbeat_pattern1[:hb_p1_N].flatten()))
hb_p2_spectrum = (2/hb_p2_N)*np.abs(sc.fft(heartbeat_pattern2[:hb_p2_N].flatten()))

hb_p1_half_spectrum = hb_p1_spectrum[:hb_p1_N//2]
hb_p2_half_spectrum = hb_p2_spectrum[:hb_p2_N//2]

hb_p1_ff = np.linspace(0, fs/2, hb_p1_N/2)
hb_p2_ff = np.linspace(0, fs/2, hb_p2_N/2)

# Registros heartbeat_pattern1 y heartbeat_pattern2 en tiempo
plt.figure(5)
plt.title('Patrón de latidos promedio')
plt.xlabel('Tiempo [seg]')
plt.ylabel('Amplitud [mV]')
plt.grid(which='both', axis='both')
plt.plot(hb_p1_tt, heartbeat_pattern1[:hb_p1_N], label='Latido Normal')
plt.plot(hb_p2_tt, heartbeat_pattern2[:hb_p2_N], label='Latido Ventricular')
axes_hdl = plt.gca()
axes_hdl.legend()
plt.show()

# Registros heartbeat_pattern1 y heartbeat_pattern2 en frecuencia
plt.figure(6)
plt.title('Espectros patrón de latidos promedio')
plt.xlabel('Frecuecnia [Hz]')
plt.ylabel('Amplitud [mV]')
plt.grid(which='both', axis='both')
plt.stem(hb_p1_ff, hb_p1_half_spectrum, linefmt='-', label='Latido Normal')
plt.stem(hb_p2_ff, hb_p2_half_spectrum, linefmt=':', label='Latido Ventricular')
axes_hdl = plt.gca()
axes_hdl.legend()
plt.show()
```

```python
# Señal qrs_detections
qrs_det_N = len(qrs_detections)

qrs_det_tt = np.linspace(0, qrs_det_N-1, qrs_det_N)

# Vector donde ocurren los latidos
plt.figure(7)
plt.title('Vector donde ocurren los latidos')
plt.xlabel('Latido [indice]')
plt.ylabel('Muestra [indice]')
plt.grid(which='both', axis='both')
plt.plot(qrs_det_tt, qrs_detections[:qrs_det_N], color='b')
plt.show()
```

```python

# Segmentos de interés #
########################
# Puede tomar como referencia los siguientes segmentos, aunque si lo considera, también
# puede incluir otros que considere importantes para evaluar el rendimiento.

zonas_con_interf_baja_frec = ( 
        np.array([12, 12.4]) *60*fs, # minutos a muestras
        np.array([15, 15.2]) *60*fs, # minutos a muestras
        )


zonas_sin_interf = ( 
        np.array([5, 5.2]) *60*fs, # minutos a muestras
        [4000, 5500], # muestras
        [10e3, 11e3], # muestras
        )

##########################################
# Acá podés generar los gráficos pedidos #
##########################################

def plot_filtered_signals(b_pb, a_pb, b_pa, a_pa, i, ecg_N_start, ecg_N_end):
    N = len(ecg_lead[ecg_N_start:ecg_N_end])
    tt = np.linspace(0, (N-1)*ts, N).flatten()

    ecg_filter_pa = sig.lfilter(b_pa, a_pa, ecg_lead.flatten())
    ecg_filter_full = sig.lfilter(b_pb, a_pb, ecg_filter_pa.flatten())
    
    ecg_ff = np.linspace(0.0, fs/2, N/2).flatten()
    
    plt.figure(i + 1)
    plt.plot(tt, ecg_lead[ecg_N_start:ecg_N_end], label='Original')
    plt.plot(tt, ecg_filter_pa[ecg_N_start:ecg_N_end], label= 'Filtrado_PA')
    plt.plot(tt, ecg_filter_full[ecg_N_start:ecg_N_end], label= 'Filtrado_PA_PB')
    plt.title('Registro ECG')
    plt.xlabel('Tiempo [seg]')
    plt.ylabel('Amplitud [mV]')
    plt.grid(which='both', axis='both')
    axes_hdl = plt.gca()
    axes_hdl.legend()
    plt.show()
    
    ecg_spectrum = (2/N)*np.abs(sc.fft(ecg_lead[ecg_N_start:ecg_N_end].flatten()))
    ecg_half_spectrum = ecg_spectrum[:N//2] 

    ecg_filter_pa_spectrum = (2/N)*np.abs(sc.fft(ecg_filter_pa[ecg_N_start:ecg_N_end].flatten()))
    ecg_filter_pa_half_spectrum = ecg_filter_pa_spectrum[:N//2] 

    ecg_filter_full_spectrum = (2/N)*np.abs(sc.fft(ecg_filter_full[ecg_N_start:ecg_N_end].flatten()))
    ecg_filter_full_half_spectrum = ecg_filter_full_spectrum[:N//2] 

    plt.figure(i + 4)

    plt.plot(ecg_ff, ecg_half_spectrum, label='Original', color='b')
    plt.plot(ecg_ff, ecg_filter_pa_half_spectrum, label='Filtrado_PA', color='g')
    plt.plot(ecg_ff, ecg_filter_full_half_spectrum, label='Filtrado_PA_PB', color='r')

    plt.title('Espectro registro ECG')
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('Amplitud [mV]')
    plt.grid(which='both', axis='both')
    axes_hdl = plt.gca()
    axes_hdl.legend()
    plt.show()
    
    return None
    
```
